```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# 1. Pembolehubah status haiwan peliharaan
status = {"kenyang": 50, "gembira": 50}

# 2. Ruang paparan visual
paparan_visual = widgets.Output()

def kemaskini_paparan():
    with paparan_visual:
        clear_output(wait=True) # Bersihkan paparan lama
        print("=" * 45)
        print(f" 🍗 Tahap Kenyang: {status['kenyang']}/100  |  ⚽ Tahap Gembira: {status['gembira']}/100 ")
        print("=" * 45)

        # Logik visual / Reaksi kucing berdasarkan status
        if status['kenyang'] >= 80 and status['gembira'] >= 80:
            print("\n        ✨ 💖 (≧◡≦) 💖 ✨")
            print("   Meoww~ Kucing anda sangat gembira dan comel!")
            print("             /\\_/\\  ")
            print("            ( o.o ) ")
            print("             > ^ <  ")
        elif status['kenyang'] <= 30:
            print("\n        💢 (╥﹏╥) 💢")
            print("   Meow... Kucing anda kelaparan! Cepat beri makan!")
            print("             /\\_/\\  ")
            print("            ( T.T ) ")
            print("             > ^ <  ")
        elif status['gembira'] <= 30:
            print("\n        💤 (︶︹︺) 💤")
            print("   Kucing anda kebosanan dan mula merajuk.")
            print("             /\\_/\\  ")
            print("            ( -.- ) ")
            print("             > ^ <  ")
        else:
            print("\n        🌟 (=^･ω･^=) 🌟")
            print("   Kucing anda sedang bersantai melihat anda.")
            print("             /\\_/\\  ")
            print("            ( ^.^ ) ")
            print("             > ^ <  ")

# 3. Logik untuk butang interaktif
def beri_makan(b):
    status['kenyang'] = min(100, status['kenyang'] + 15)
    kemaskini_paparan()

def bermain(b):
    status['gembira'] = min(100, status['gembira'] + 15)
    status['kenyang'] = max(0, status['kenyang'] - 5) # Bermain buat kucing lapar sedikit
    kemaskini_paparan()

def biarkan(b):
    status['gembira'] = max(0, status['gembira'] - 10) # Kucing bosan kalau dibiarkan
    status['kenyang'] = max(0, status['kenyang'] - 10)
    kemaskini_paparan()

# 4. Reka bentuk Butang Interaktif
btn_makan = widgets.Button(description='🍗 Beri Makan', button_style='success')
btn_main = widgets.Button(description='⚽ Bermain', button_style='info')
btn_biar = widgets.Button(description='⏳ Biarkan Saja', button_style='warning')

# 5. Sambungkan butang dengan fungsinya
btn_makan.on_click(beri_makan)
btn_main.on_click(bermain)
btn_biar.on_click(biarkan)

# Susun butang dengan kemas dalam satu barisan mendatar
menu_butang = widgets.HBox([btn_makan, btn_main, btn_biar])

# 6. Paparkan antaramuka ke skrin Colab
display(paparan_visual)
display(menu_butang)

# Panggil fungsi kemaskini untuk paparan pertama kali
kemaskini_paparan()
```
